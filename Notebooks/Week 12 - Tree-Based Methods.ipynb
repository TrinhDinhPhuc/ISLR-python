```python
# %load ../standard_import.txt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_rows', 200) 
pd.set_option('display.max_columns', 200) 
%matplotlib inline
plt.style.use('seaborn-white')
```

```python
df = pd.read_csv("Data/Hitters.csv").dropna()
df.info()
```

```python
X = df[['Years','Hits']].values
y = np.log(df.Salary.values) 
"""
The numpy. log() is a mathematical function that is used to calculate the natural logarithm
of x(x belongs to all the input array elements)
"""
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(11,4))
ax1.hist(df.Salary.values)
ax1.set_xlabel("Salary")
ax1.set_ylabel("Hits")
ax2.hist(y)
ax2.set_xlabel("Log(salary)")
ax2.set_ylabel("Hits")
```

```python
import pydot
from IPython.display import Image
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.externals.six import StringIO  
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier, export_graphviz
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier, BaggingRegressor, RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error,confusion_matrix, classification_report
# This function creates images of tree models using pydot
def print_tree(estimator, features, class_names=None, filled=True):
    tree = estimator
    names = features
    color = filled
    classn = class_names
    
    dot_data = StringIO()
    export_graphviz(estimator, out_file=dot_data, feature_names=features, class_names=classn, filled=filled)
    graph = pydot.graph_from_dot_data(dot_data.getvalue())
    return(graph)
%matplotlib inline
plt.style.use('seaborn-white')
```

## Maximum Leaf Nodes = 3

```python
regr = DecisionTreeRegressor(max_leaf_nodes=3)
regr.fit(X, y)
graph, = print_tree(regr, features=['Years', 'Hits'])
Image(graph.create_png())
```

## Maximum Leaf Nodes = 4

```python
regr = DecisionTreeRegressor(max_leaf_nodes=4)
regr.fit(X, y)
graph, = print_tree(regr, features=['Years', 'Hits'])
Image(graph.create_png())
```

## Maximum Leaf Nodes = 5

```python
regr = DecisionTreeRegressor(max_leaf_nodes=15)
regr.fit(X, y)
graph, = print_tree(regr, features=['Years', 'Hits'])
Image(graph.create_png())
```
